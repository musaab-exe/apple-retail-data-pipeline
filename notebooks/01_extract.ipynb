```python
import sys
sys.path.append('..')

from utils.spark_session import get_spark
```

### Starting Spark Session ###

```python
spark = get_spark()
spark
```

### 1. Reading the Raw CSVs 

```python
category_df = spark.read.csv(
    '../data/raw/category.csv', 
    header=True, 
    inferSchema= True
    )

products_df = spark.read.csv(
    '../data/raw/products.csv',
    header=True, 
    inferSchema= True
    )

sales_df = spark.read.csv(
    '../data/raw/sales.csv',
    header=True, 
    inferSchema= True
    )

stores_df = spark.read.csv(
    '../data/raw/stores.csv',
    header=True, 
    inferSchema= True
    )

warranty_df = spark.read.csv(
    '../data/raw/warranty.csv',
    header=True, 
    inferSchema= True
    )
```

### 2. Uploading to Bronze ###

```python
category_df.write.mode('overwrite').parquet('../data/bronze_ingested/category')
products_df.write.mode('overwrite').parquet('../data/bronze_ingested/products')
sales_df.write.mode('overwrite').parquet('../data/bronze_ingested/sales')
stores_df.write.mode('overwrite').parquet('../data/bronze_ingested/store')
warranty_df.write.mode('overwrite').parquet('../data/bronze_ingested/warranty')
```
